1. Python Program to find sum of an Array

```python
def sum_arr(arr):
    sum = 0
    for i in range(len(arr)):
        sum += arr[i]
        
    return sum

li = [1,2,3,4,5,6]
result = sum_arr(li)

print("The sum of the Array is: ", result)
```

2. Python Program to find largest number in an Array

```python
def find_largest(arr2):
    large = 0
    for i in range(len(arr2)-1):
        if arr2[i] < arr2[i+1]:
            large = arr2[i+1]
        
    return large
li2 = [4,5,63,44,75,0]
result2 = find_largest(li2)
print("The Largest number in the array is : ", result2)
```

3. Python Program for Array rotation

```python
arr3 = [1,2,3,4,5,6]
limit = 2

def print_array(arr):
    li = []
    for i in range(len(arr)):
        li.append((arr[i]))
    return li

def rotate_array(arr, limi):
    for i in range(0, limi):
        temp = arr[0]
        for j in range(0, len(arr)-1):
            arr[j] = arr[j+1]
        arr[len(arr)-1]= temp
        
    return arr3


before = print_array(arr3)
print("Array before Rotation: ", before)
res = rotate_array(arr3, limit)
print("Array after rotation: ", res)
```

4. Python Program to Split the array and add the first part to the end

```python
arr4 = [1,2,3,4,5,6]
shift = 2

def print_array(arr):
    li = []
    for i in range(len(arr)):
        li.append((arr[i]))
    return li

def split_array(arr, limi):
    for i in range(0, limi):
        temp = arr[0]
        for j in range(0, len(arr)-1):
            arr[j] = arr[j+1]
        arr[len(arr)-1]= temp
        
    return arr3


before = print_array(arr3)
print("Array before split: ", before)
res = split_array(arr4, shift)
print("Array after split: ", res)
```

5.  Python Program to check if given array is Monotonic

```python
def is_monotonic(array):
    decreasing = False
    if array[0] > array[1]:
        decreasing = True
    
    for i in range(len(array)-1):
        if(array[i] < array[i+1] and decreasing or array[i] > array[i+1] and not decreasing):
            return False
    return True
    
print(is_monotonic([1,2,8,4,3,5]))
print(is_monotonic([1,2,3,4,5,6,7]))
```
