```python
def calculate_lcm(a,b):
    if a > b:
        greater = a
    else:
        greater = b

    while(True):
        if(greater % a == 0) and (greater % b == 0):
            lcm = greater
            break
        greater+=1
    return lcm

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the Second number: "))

print("First Number: ", num1)
print("Second Number:", num2)
print("The L.C.M is :",calculate_lcm(num1,num2))
```

```python
def calculate_hcf(x, y):    
    if x > y:  
        smaller = y  
    else:  
        smaller = x  
    for i in range(1,smaller + 1):  
        if((x % i == 0) and (y % i == 0)):  
            hcf = i  
    return hcf  
 
num1 = int(input("Enter the first number: "))  
num2 = int(input("Enter the second number: "))  

print("The H.C.F. of", num1,"and", num2,"is", calculate_hcf(num1, num2)) 
```

```python
def decimal_to_binary(deci):
    d_number = int(deci)
    print(deci, " in binary is: ", bin(d_number))

number = int(input("Enter the Decimal Number: "))

decimal_to_binary(number)

```

```python
def OctToHex(o):
    return hex(int(o, 8))

print("Enter Octal Number: ", end="")
onum = input()

hnum = OctToHex(onum)
print("\nEquivalent Hexadecimal Value =", hnum[2:].upper())
```

```python
def add(a,b):
    print(a + b) 

def subtract(a,b):
    print(a-b) 

def divide(a,b):
    print(a/b) 

def mul(a,b):
    print(a * b) 

x = int(input("Enter the first number: "))
y = int(input("Enter the second number: "))
print("Press 1 to add")
print("Press 2 to subtract")
print("Press 3 to divide")
print("Press 4 to Multiply")

choice = int(input("Enter your choice: "))

if(choice == 1):
    add(x,y)
elif choice == 2:
    subtract(x,y)
elif choice == 3:
    divide(x,y)
elif choice == 4:
    mul(x,y)
```
